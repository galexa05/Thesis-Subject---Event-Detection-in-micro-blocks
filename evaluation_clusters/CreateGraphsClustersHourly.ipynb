```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ClusterSpecsHourly.csv',header='infer')
print(df.shape[0])
df.head(1)
```

```python
# create a figure and axis 
fig, ax = plt.subplots(figsize=(10,10)) 
# count the occurrence of each class 
data = df['Purity'].value_counts() 
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('Purity of Clusters') 
ax.set_xlabel('Purity') 
ax.set_ylabel('Frequency')
```

```python
# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = df['Predicted_Event'].value_counts() 
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('Purity of Clusters') 
ax.set_xlabel('Event') 
ax.set_ylabel('Number of Cluster')
```

```python
new_df = df.loc[df['Predicted_Event']!=-1]
```

```python
# create a figure and axis 
fig, ax = plt.subplots(figsize=(10,10)) 
# count the occurrence of each class 
data = new_df['Predicted_Event'].value_counts() 
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('Purity of Clusters') 
ax.set_xlabel('Event') 
ax.set_ylabel('Number of Cluster')
```
